```python
import matplotlib.pyplot as plt
import numpy as np
Order_array = np.array(df2["Order Date"])
Order_array
Sales_array = np.array(df2["Sales"])
Sales_array 
plt.plot(Order_array, Sales_array)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

Category_array = np.array(df2['Category'])
Category_array
Sales_array = np.array(df2["Sales"])
Sales_array 
plt.bar(Category_array, Sales_array)
plt.title("Total amount of Sales per category")
plt.xlabel("Category")
plt.ylabel("Total Sales")
plt.show()

number_of_sales_per_category = df2['Category'].value_counts()
plt.bar(number_of_sales_per_category.index, number_of_sales_per_category.values)
plt.title("Number of sales per category")
plt.xlabel('Category')
plt.ylabel('Number of sales')
plt.show()

sales_per_category = df2.groupby("Category")["Sales"].sum()
print(sales_per_category)
```
